# Task
Analyze the relationship between GDP per capita and happiness score using the World Bank and World Happiness Report datasets. Visualize the relationship using scatter plots, calculate the correlation coefficient, and segment the analysis by region. Summarize the findings in a report.

## Load and prepare data

### Subtask:
Load the dataset and prepare it for analysis by selecting the relevant columns: 'Country', 'Region', 'Economy (GDP per Capita)', and 'Happiness Score'.


**Reasoning**:
The first step is to load the dataset and select the relevant columns as specified in the instructions.



```python
import pandas as pd

# Load the dataset
df = pd.read_csv('2015.csv')

# Select the relevant columns
df_numerical = df[['Country', 'Region', 'Economy (GDP per Capita)', 'Happiness Score']]

# Display the first few rows of the new DataFrame
display(df_numerical.head())
```

## Visualize gdp vs happiness

### Subtask:
Create a scatter plot to visualize the relationship between 'Economy (GDP per Capita)' and 'Happiness Score'.


**Reasoning**:
Create a scatter plot to visualize the relationship between 'Economy (GDP per Capita)' and 'Happiness Score'.



```python
import matplotlib.pyplot as plt

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df_numerical['Economy (GDP per Capita)'], df_numerical['Happiness Score'])

# Add labels and title
plt.xlabel('Economy (GDP per Capita)')
plt.ylabel('Happiness Score')
plt.title('Relationship between Economy (GDP per Capita) and Happiness Score')

# Display the plot
plt.show()
```

**Reasoning**:
The scatter plot has been created. The next step is to calculate the correlation coefficient between 'Economy (GDP per Capita)' and 'Happiness Score'.



```python
# Calculate the correlation coefficient
correlation_coefficient = df_numerical['Economy (GDP per Capita)'].corr(df_numerical['Happiness Score'])

# Print the correlation coefficient
print(f"The correlation coefficient between Economy (GDP per Capita) and Happiness Score is: {correlation_coefficient:.2f}")
```

**Reasoning**:
The scatter plot and correlation coefficient have been generated. The next step is to visualize the relationship between 'Economy (GDP per Capita)' and 'Happiness Score' for each region using scatter plots.



```python
# Get unique regions
regions = df_numerical['Region'].unique()

# Create a scatter plot for each region
plt.figure(figsize=(15, 10))
for region in regions:
    df_region = df_numerical[df_numerical['Region'] == region]
    plt.scatter(df_region['Economy (GDP per Capita)'], df_region['Happiness Score'], label=region, alpha=0.7)

# Add labels, title, and legend
plt.xlabel('Economy (GDP per Capita)')
plt.ylabel('Happiness Score')
plt.title('Relationship between Economy (GDP per Capita) and Happiness Score by Region')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()

# Display the plot
plt.show()
```

## Analyze correlation

### Subtask:
Calculate and display the correlation coefficient between 'Economy (GDP per Capita)' and 'Happiness Score'.


**Reasoning**:
Calculate the correlation coefficient between the 'Economy (GDP per Capita)' and 'Happiness Score' columns and print it formatted to two decimal places.



```python
# Calculate the correlation coefficient
correlation_coefficient = df_numerical['Economy (GDP per Capita)'].corr(df_numerical['Happiness Score'])

# Print the correlation coefficient
print(f"The correlation coefficient between Economy (GDP per Capita) and Happiness Score is: {correlation_coefficient:.2f}")
```

## Segment by region

### Subtask:
Segment the data by region and analyze the relationship between GDP and Happiness within each region.


**Reasoning**:
Group the DataFrame by region and calculate the correlation coefficient for the specified columns within each group. Store the results in a dictionary.



```python
# Group by region and calculate the correlation within each region
regional_correlation = df_numerical.groupby('Region')[['Economy (GDP per Capita)', 'Happiness Score']].corr().unstack().iloc[:,1]

# Store the results in a dictionary
regional_correlation_dict = regional_correlation.to_dict()

# Display the regional correlation coefficients
print("Correlation between Economy (GDP per Capita) and Happiness Score by Region:")
for region, correlation in regional_correlation_dict.items():
    print(f"{region}: {correlation:.2f}")
```
